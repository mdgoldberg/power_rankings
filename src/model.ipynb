```python
from collections import defaultdict

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import linear_model

import power_rankings
import rank_functions
```

```python
dfs = [power_rankings.get_inputs('../html/home{}.htm'.format(yr)) for yr in range(2012, 2017)]
```

```python
prs = [[rank_functions.expected_win_pct(df[df.week < week], 1, week - 1)
        for week in range(2, rank_functions.most_recent_week(df) + 1)]
       for df in dfs]
```

```python
def agg_by_team(pr):
    by_team = defaultdict(list)
    for pr in this_year:
        for team, val in pr.items():
            by_team[team].append(val)
    return by_team 
```

```python
def plot_by_team(pr):    
    by_team = agg_by_team(pr)
    fig, axes = plt.subplots(3, 4, figsize=(15, 10))
    for ax, (team, ranks) in zip(np.ravel(axes), by_team.items()):
        ax.plot(range(1, len(ranks)+1), ranks, 'o-')
        ax.set_title(team)
        ax.set_ylim([0, 1])
    plt.tight_layout()
    plt.show()

this_year = prs[-1]
plot_by_team(this_year)
```

```python
# TODO build training data and model
# columns: home ranking, away ranking, home record (wins), away record (wins), home PF, away PF, home_won

def data_from_game(row, pr):
    # rankings
    week_pr = pr[row['week'] - 2]
    team_rank = week_pr[row['team']]
    opp_rank = week_pr[row['opponent']]
    log5 = (team_rank - team_rank*opp_rank) / (team_rank + opp_rank - 2*team_rank*opp_rank)
    return pd.Series({'prob': log5, 'team_rank': team_rank, 'opp_rank': opp_rank})
```

```python
data_dfs = [df.ix[(df.week >= 2) & (df.week < rank_functions.most_recent_week(df) + 1)] for df in dfs]
data_dfs = [pd.concat((df, df.apply(data_from_game, pr=pr, axis=1)), axis=1) for pr, df in zip(prs, data_dfs)]
data_df = pd.concat(data_dfs)
```
